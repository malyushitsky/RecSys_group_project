## Kaggle

### metadata

```python
pd.set_option('display.max_columns', None)

path = 'movies_metadata.csv'
kgl_movies_metadata = pd.read_csv(path)

kgl_movies_metadata['release_year'] = kgl_movies_metadata['release_date'].str[:4]
kgl_movies_metadata = kgl_movies_metadata.loc[~kgl_movies_metadata.duplicated(['original_title', 'release_date'])]

kgl_movies_metadata.head()
```

```python
kgl_movies_to_merge = kgl_movies_metadata[['imdb_id', 'original_title', 'release_year']].copy()
kgl_movies_to_merge.original_title = kgl_movies_to_merge.original_title.str.lower().str.strip()
# kgl_movies_to_merge.columns = ['kgl_id', 'kgl_original_title', 'release_year']
kgl_movies_to_merge
```

```python
merged_movies = movie_metadata_to_merge.merge(kgl_movies_to_merge,
                                     how='inner',
                                     left_on=['Movie name', 'release_year'],
                                     right_on=['original_title', 'release_year'])
merged_movies = merged_movies.loc[~merged_movies.imdb_id.isna()]
merged_movies
```

```python
merged_movies.imdb_id.nunique()
```

```python
import pandas as pd
path = 'ratings.csv'
ratings_big_raw = pd.read_csv(path)

ratings_big_raw.head()
```

```python
ratings_big_raw.shape, ratings_big_raw.movieId.nunique(), ratings_big_raw.userId.nunique()
```

```python
path = 'links.csv'
links_big = pd.read_csv(path)

links_big.head()
```

```python
small_cols = ['userId', 'rating', 'timestamp', 'imdbId']
ratings_big = ratings_big_raw.merge(links_big, how='inner', on='movieId')[small_cols]
ratings_big.imdbId = \
ratings_big.imdbId.astype(str).apply(lambda x: x if pd.isnull(x) else 'tt' + x.rjust(7, '0'))

ratings_big
```

```python
tt = ratings_big.groupby(['userId', 'imdbId'], as_index=False).agg({'rating':'count'})
tt.loc[tt.rating > 1]
```
